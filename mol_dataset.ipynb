# NP Atlas

```python
from rdkit import Chem

# Open the .sdf file
sdf_file = '../Data/NPAtlas/NPAtlas_2024_09.sdf'
suppl = Chem.SDMolSupplier(sdf_file)

npatlas_smiles = []

# Iterate over the molecules in the .sdf file
for mol in suppl:
    if mol is not None:
        # Do something with the molecule
        # For example, print the molecule's SMILES representation
        npatlas_smiles.append(mol.GetProp('compound_smiles'))

print(f'\nExtracted {len(npatlas_smiles)} SMILES from NPAtlas.')
```
